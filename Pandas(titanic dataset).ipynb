MERGING (It mainly takes place on similar columns )

```python
fd1=pd.DataFrame({"key1":[1,8,3,9,5,6],
                 "key2":[4,5,6,7,8,9],
                 "key3":[11,22,33,44,55,76]})
fd2=pd.DataFrame({"key1":[1,2,3,4,5,6],
                 "key4":[87,4,6,3,1,9],
                 "key5":[3,56,5,6,6,7]})
```

```python
pd.merge(fd1,fd2)
```

```python
pd.merge(fd1,fd2,how="outer")
```

```python
pd.merge(fd1,fd2,how="left")
```

```python
pd.merge(fd1,fd2,how="right")
```

JOIN OPERATION (Mainly takes place of index

```python
fd1=pd.DataFrame({"key1":[1,8,3,9,5,6],
                 "key2":[4,5,6,7,8,9],
                 "key3":[11,22,33,44,55,76]})
fd2=pd.DataFrame({"key11":[1,2,3,4,5,6],
                 "key4":[87,4,6,3,1,9],
                 "key5":[3,56,5,6,6,7]})
```

```python
fd1.join(fd2)
```

```python
fd1=pd.DataFrame({"key1":[1,8,3,9,5,6],
                 "key2":[4,5,6,7,8,9],
                 "key3":[11,22,33,44,55,76]},
                  index=["a","b","c","d","e","f"])
fd2=pd.DataFrame({"key11":[1,2,3,4,5,6],
                 "key4":[87,4,6,3,1,9],
                 "key5":[3,56,5,6,6,7]},
                index=["a","f","b","h","i","m"])
```

```python
fd1.join(fd2)
```
